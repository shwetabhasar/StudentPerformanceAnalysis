```python
import pandas as pd

data = pd.read_csv('student_data.csv')
data
```

```python
import pandas as pd
data['Average'] = data[['Maths','Science','English']].mean(axis=1)
data
```

```python
print(data.columns.tolist())
```

```python
data.columns = data.columns.str.strip()
```

```python
data['Average'] = data[['Maths','Science','English']].mean(axis=1)
data
```

```python
topper = data.loc[data['Average'].idxmax()]
print(topper)
```

```python
failed = data[data['Average'] < 40]
print(failed)
```

```python
data.sort_values(by='Average', ascending=False)
```

```python
# Grade system
data['Grade'] = data['Average'].apply(lambda x: 'A' if x>=80 else 'B' if x>=60 else 'C')

# Topper
print(data.loc[data['Average'].idxmax()])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure()
plt.bar(data['Name'], data['Average'])
plt.title("Average Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Average Marks")
plt.xticks(rotation=45)
plt.show()
```

```python
data[['Maths','Science','English']].mean().plot(kind='bar')
plt.title("Average Marks per Subject")
plt.show()
```

```python
plt.figure()
plt.scatter(data['Attendance'], data['Average'])
plt.title("Attendance vs Average Marks")
plt.xlabel("Attendance")
plt.ylabel("Average Marks")
plt.show()
```

```python
data.to_csv("final_student_data.csv", index=False)
```
